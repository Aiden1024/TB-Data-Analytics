### Describing Numerical Variables

```python
pd.set_option('display.width',100)
pd.set_option('precision',5)

sales.describe()
```

```python
# Upper Fence = Q3 + 1.5*IQR
# Lower Fence = Q1 - 1.5*IQR
Q1 = 2
Q3 = 12
IQR = Q3 - Q1

UF = Q3 + 1.5*IQR
LF = Q1 - 1.5*IQR

print(UF)
print(LF)
```

```python
sales_no_outliers = sales[sales['Qty'] < UF]
sales_no_outliers = sales_no_outliers[sales_no_outliers['Qty'] > LF]
sales_no_outliers.head()
```

```python
sales.describe(percentiles=[.025,.25,.5,.75,.9725])
```

```python
sales.mode()
```

```python
from scipy.stats import zscore
```

```python
standardized = pd.DataFrame(sales.groupby('CusID')['Qty'].transform(zscore))

outliers = ((standardized['Qty'] < -3) | (standardized['Qty'] > 3))

sales_outliers = sales[outliers]
print(sales_outliers.head())
```

```python
sales_outliers.describe()
```

```python
mean_Qty = sales['Qty'].mean()
SD_Qty = sales['Qty'].std()

sales_no_outliers = sales[np.abs(sales['Qty']) < (mean_Qty + 3*SD_Qty)]
sales_no_outliers.head()
```

```python
sales_no_outliers.describe()
```
